```python
# traverse all the values and fill the null values based on the class
def func(var):

    a = var[0]
    b = var[0]
    
    if pd.isnull(a):
        if b ==1:
            return 38
        elif b==2:
            return 30
        else:
            return 25
    else:
        return a
```

```python
df['Age'] =df[['Age', 'Pclass']].apply(func, axis=1)
```
